# Conclusion
## NumPy Datatypes Update
Compared to the regular naive implementation, we can see that a sligth improvement of the runtime can be made by using the NumPy float64 datatype. This is funny since that is the datatype which is usually default, so it seems a bit arbitrary.

## Dask Implementation
With the regular NumPy version, we see a runtime of 33.44 seconds, whereas in the best case dask implementation we have a runtime of 15.48 seconds, and thus over a 50% increase in performance.
However, we can also see that the chunksize plays a big role. $(1500, 1)$ apparently yields a runtime of 34.57 seconds, which is even worse than that of the regular NumPy version.
